```python
import pandas as pd;
import numpy as np;
import math;

filePath = "D:\\Applied Hands-on\\ENGR 6412 Assignment 2\\Data\\map2023DAT.dat"

dfMap = pd.read_csv(filePath, delimiter='\t')
dfMap.head(100)
```

```python
# Function to calculate Euclidean distance
def euclidean_distance(x1, y1, x2, y2):
    return ((x1 - x2)**2 + (y1 - y2)**2)**0.5

# Function to find nearby 1s and their distances (in all directions)
def find_nearby_ones(dataset, x, y):
    closest_distance = float('inf')
    closest_1_position = None

    # Iterate through angles from 0 to 360 degrees
    for angle in range(0, 360):
        # Convert angle to radians
        radians = angle * 3.14159 / 180.0

        # Calculate x and y components using trigonometry
        dx = round(math.cos(radians))
        dy = round(math.sin(radians))

        # Check cells along the direction defined by dx and dy
        i, j = x, y
        while 0 <= i < len(dataset) and 0 <= j < len(dataset.columns):
            if dataset.iloc[i, j] == 1 and (i != x or j != y):
                distance = euclidean_distance(x, y, i, j)
                if distance < closest_distance:
                    closest_distance = distance
                    closest_1_position = (i, j)
                break  # Break loop if a '1' is found in this direction

            # Move along the direction
            i += dx
            j += dy

    return closest_1_position, closest_distance
```

```python
def pHit_calculator(closest_distance):
    distanceToObstacle = 0

    for i in range(1, closest_distance + 1):
        distanceToObstacle += math.pow(i * 4, 2)  # Ensure 4 is in centimeters
    distanceToObstacle = round(math.sqrt(distanceToObstacle), 2)

    variance_cm = 25  # Variance given in centimeters

    pHit = (1 / (math.sqrt(2 * math.pi * variance_cm))) * math.exp(-(math.pow(distanceToObstacle, 2)) / (2 * variance_cm))
    return pHit
```

```python
for i in range(0, int(dfMap.shape[0])):
    for j in range(0, int(dfMap.shape[1])):
        if dfMap.iloc[i, j] == 1:
            dfLikelihood.at[i , j] = 0.1
            continue
        else:
            closest_1_position, closest_distance = find_nearby_ones(dfMap, i, j)
            print(closest_1_position)
            print(closest_distance)
            if closest_distance*4 >= 20:
              dfLikelihood.at[i, j] = 0
            else:
              dfLikelihood.at[i, j] = pHit_calculator(int(closest_distance))
        print(dfMap.iloc[i, j])
        print("Row" + str(i))
```
